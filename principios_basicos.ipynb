Iniciaremos con la declaración de una función muy sencilla, la función ```print()```. Esta función imprime el valor que encuentra entre los paréntesis ("texto" o número).

Además de los parentésis, agregamos comilla doble o sencilla para indicar que lo que está entre comillas, no es código.

# Función input()

La función ```input()``` es para solicitar al usuario que introduzca algo a través del teclado (entrada estándar).
